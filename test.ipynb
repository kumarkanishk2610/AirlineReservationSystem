```python
import tkinter as tk
from tkinter import ttk

def create_flight_details_widget(parent, flight_details):
    # Create a frame to hold the flight details
    details_frame = ttk.Frame(parent, padding=(10, 10))
    details_frame.grid(sticky="nsew")

    # Define flight details labels
    labels = [
        "Airline", "Flight No.", "Source", "Destination",
        "Departure Date", "Departure Time", "Arrival Date",
        "Arrival Time", "Duration", "Price", "Flight Status",
        "Available Seats", "Booking Class", "Aircraft Name", "Cabin Configuration"
    ]

    # Create and grid flight details labels
    for i, label_text in enumerate(labels):
        label = ttk.Label(details_frame, text=label_text, font=("Helvetica", 10, "bold"))
        label.grid(row=i, column=0, sticky="w", padx=(0, 5), pady=(0, 5))

        value = ttk.Label(details_frame, text=flight_details.get(label_text, ""), font=("Helvetica", 10))
        value.grid(row=i, column=1, sticky="w", padx=(0, 5), pady=(0, 5))

    # Configure grid weights to make labels expandable
    details_frame.columnconfigure(0, weight=1)
    details_frame.columnconfigure(1, weight=3)

# Sample flight details
flight_details = {
    "Airline": "Example Airlines",
    "Flight No.": "EA123",
    "Source": "JFK",
    "Destination": "LAX",
    "Departure Date": "2024-04-15",
    "Departure Time": "08:00",
    "Arrival Date": "2024-04-15",
    "Arrival Time": "10:30",
    "Duration": "2h 30m",
    "Price": "$200",
    "Flight Status": "On Time",
    "Available Seats": "100",
    "Booking Class": "Economy",
    "Aircraft Name": "Boeing 737",
    "Cabin Configuration": "2-3-2"
}

# Create the main window
root = tk.Tk()
root.title("Flight Details Widget")

# Create the flight details widget
create_flight_details_widget(root, flight_details)

# Start the tkinter event loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

# Create the main window
root = tk.Tk()
root.title("Flight Information")

# Create the search_frame (assuming it's already created)
search_frame = tk.Frame(root, bg="#f2f2f2", relief=tk.GROOVE)
search_frame.pack(fill=tk.BOTH, expand=True)

# Create the output_frame
output_frame = tk.Frame(search_frame, bg="#f2f2f2")
output_frame.pack(side="right", fill=tk.BOTH, expand=True, padx=0, pady=(0, 0))

# Function to create flight information card
def create_flight_card(parent, flight_info):
    # Create the frame for the card
    card_frame = tk.Frame(parent, bg="white", bd=1, relief=tk.RAISED)
    card_frame.pack(fill=tk.X, padx=10, pady=10)

    # Airline and Flight Number section
    airline_flight_frame = tk.Frame(card_frame, bg="white")
    airline_flight_frame.pack(fill=tk.X)

    # Airline label
    airline_label = tk.Label(airline_flight_frame, text="Airline: ", bg="white")
    airline_label.pack(side=tk.LEFT, padx=(5, 0), pady=2)

    # Airline name
    airline_name = tk.Label(airline_flight_frame, text=flight_info["Airline"], bg="white")
    airline_name.pack(side=tk.LEFT, padx=(0, 10), pady=2)

    # Flight Number label
    flight_number_label = tk.Label(airline_flight_frame, text="Flight Number: ", bg="white")
    flight_number_label.pack(side=tk.LEFT, padx=(0, 0), pady=2)

    # Flight Number
    flight_number = tk.Label(airline_flight_frame, text=flight_info["Flight No."], bg="white")
    flight_number.pack(side=tk.LEFT, padx=(0, 5), pady=2)

    # Source and Destination section
    source_dest_frame = tk.Frame(card_frame, bg="white")
    source_dest_frame.pack(fill=tk.X)

    # Source label
    source_label = tk.Label(source_dest_frame, text="Source: ", bg="white")
    source_label.pack(side=tk.LEFT, padx=(5, 0), pady=2)

    # Source
    source = tk.Label(source_dest_frame, text=flight_info["Source"], bg="white")
    source.pack(side=tk.LEFT, padx=(0, 10), pady=2)

    # Destination label
    destination_label = tk.Label(source_dest_frame, text="Destination: ", bg="white")
    destination_label.pack(side=tk.LEFT, padx=(0, 0), pady=2)

    # Destination
    destination = tk.Label(source_dest_frame, text=flight_info["Destination"], bg="white")
    destination.pack(side=tk.LEFT, padx=(0, 5), pady=2)

    # Departure and Arrival section
    dep_arr_frame = tk.Frame(card_frame, bg="white")
    dep_arr_frame.pack(fill=tk.X)

    # Departure date and time label
    dep_date_time_label = tk.Label(dep_arr_frame, text="Departure: ", bg="white")
    dep_date_time_label.pack(side=tk.LEFT, padx=(5, 0), pady=2)

    # Departure date and time
    dep_date_time = tk.Label(dep_arr_frame, text=f"{flight_info['DepDt']} | {flight_info['DepTime']}", bg="white")
    dep_date_time.pack(side=tk.LEFT, padx=(0, 10), pady=2)

    # Arrival date and time label
    arr_date_time_label = tk.Label(dep_arr_frame, text="Arrival: ", bg="white")
    arr_date_time_label.pack(side=tk.LEFT, padx=(0, 0), pady=2)

    # Arrival date and time
    arr_date_time = tk.Label(dep_arr_frame, text=f"{flight_info['ArrDt']} | {flight_info['ArrTime']}", bg="white")
    arr_date_time.pack(side=tk.LEFT, padx=(0, 5), pady=2)

    # Add flight information labels to the card
    labels = ["Duration", "Price", "FlightStatus", "AvailableSeats", "BookingClass", "AircraftName", "CabinCfg"]
    for label in labels:
        label_text = label + ": " + flight_info[label]  # Assuming flight_info is a dictionary containing flight details
        tk.Label(card_frame, text=label_text, bg="white").pack(anchor="w", padx=5, pady=2, fill=tk.X)

# Sample flight information
flight_info = {
    "Airline": "Example Airlines",
    "Flight No.": "EA123",
    "Source": "JFK",
    "Destination": "LAX",
    "DepDt": "2024-04-15",
    "DepTime": "08:00",
    "ArrDt": "2024-04-15",
    "ArrTime": "10:30",
    "Duration": "2h 30m",
    "Price": "$200",
    "FlightStatus": "On Time",
    "AvailableSeats": "100",
    "BookingClass": "Economy",
    "AircraftName": "Boeing 737",
    "CabinCfg": "2-3-2"
}

# Create the main window
root = tk.Tk()
root.title("Flight Information")

# Create the output_frame
output_frame = tk.Frame(root, bg="#f2f2f2")
output_frame.pack(side="right", fill=tk.BOTH, expand=True, padx=0, pady=(0, 0))

# Create flight information cards
create_flight_card(output_frame, flight_info)

# Start the tkinter event loop
root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

def parse_flight_data(data):
    """
    Parse the flight data into a list of dictionaries.
    """
    flight_list = []
    for flight_info in data:
        flight_dict = {
            "Airline": flight_info[0],
            "Flight No.": flight_info[1],
            "Source": flight_info[2],
            "Destination": flight_info[3],
            "Departure Date": flight_info[4],
            "Departure Time": flight_info[5],
            "Arrival Date": flight_info[6],
            "Arrival Time": flight_info[7],
            "Duration": flight_info[8],
            "Price": 'â‚¹' + str(flight_info[9]),
            "Flight Status": flight_info[10],
            "Available Seats": flight_info[11],
            "Booking Class": flight_info[12],
            "Aircraft Name": flight_info[13],
            "Cabin Configuration": flight_info[14]
        }
        flight_list.append(flight_dict)
    return flight_list

def display_flight_table(parent, flight_data):
    """
    Create and display the flight table in the specified parent widget.
    """
    # Create the treeview widget
    tree = ttk.Treeview(parent)

    # Define columns
    columns = list(flight_data[0].keys())
    tree["columns"] = tuple(columns)
    tree.heading("#0", text="Index")
    for column in columns:
        tree.heading(column, text=column)

    # Set column widths
    column_widths = {
        "Airline": 150,
        "Flight No.": 100,
        "Source": 70,
        "Destination": 70,
        "Departure Date": 100,
        "Departure Time": 100,
        "Arrival Date": 100,
        "Arrival Time": 100,
        "Duration": 70,
        "Price": 70,
        "Flight Status": 100,
        "Available Seats": 70,
        "Booking Class": 100,
        "Aircraft Name": 150,
        "Cabin Configuration": 150
    }
    for column, width in column_widths.items():
        tree.column(column, width=width)

    # Insert data rows
    for i, flight in enumerate(flight_data, start=1):
        tree.insert("", "end", text=str(i), values=tuple(flight.values()))

    # Pack the treeview widget
    tree.pack(fill="both", expand=True)

# Sample flight data
data = [('AIX Connect', 'IAD6226', 'BLR', 'BOM', '2024-04-14', '00:50', '2024-04-14', '01:10', '00:20', 56616, 'DELAYED', 24, 'Economy', 'Boeing 737', '3+3'), ('Air India Express', 'AXB3096', 'BLR', 'BOM', '2024-04-14', '03:10', '2024-04-14', '06:10', '03:00', 13256, 'ON TIME', 32, 'Economy', 'Boeing 747', '3+4+3'), ('Vistara', 'UK 5785', 'BLR', 'BOM', '2024-04-14', '04:00', '2024-04-14', '07:15', '03:15', 12212, 'DELAYED', 54, 'Economy', 'Boeing 787', '3+3+3'), ('Vistara', 'UK 8286', 'BLR', 'BOM', '2024-04-14', '05:15', '2024-04-14', '07:40', '02:25', 49737, 'ON TIME', 62, 'Economy', 'Airbus A321neo', '3+3'), ('Indigo', '6E 9170', 'BLR', 'BOM', '2024-04-14', '05:40', '2024-04-14', '07:40', '02:00', 77415, 'ON TIME', 45, 'Economy', 'Boeing 747', '3+4+3'), ('SpiceJet', 'SG 6790', 'BLR', 'BOM', '2024-04-14', '10:20', '2024-04-14', '12:45', '02:25', 79669, 'ON TIME', 15, 'Economy', 'Boeing 787', '3+3+3'), ('Air India Express', 'AXB9314', 'BLR', 'BOM', '2024-04-14', '10:55', '2024-04-14', '14:20', '03:25', 7774, 'ON TIME', 37, 'Economy', 'Boeing 737', '3+3'), ('SpiceJet', 'SG 8704', 'BLR', 'BOM', '2024-04-14', '16:20', '2024-04-14', '18:15', '01:55', 4193, 'ON TIME', 38, 'Economy', 'Airbus A320neo', '3+3'), ('AIX Connect', 'IAD3777', 'BLR', 'BOM', '2024-04-14', '23:45', '2024-04-15', '03:35', '03:50', 73818, 'ON TIME', 61, 'Economy', 'Boeing 747', '3+4+3')]

# Parse the flight data
flight_data = parse_flight_data(data)

# Create the main window
root = tk.Tk()
root.title("Flight Information Table")

# Display the flight table
display_flight_table(root, flight_data)

# Start the tkinter event loop
root.mainloop()
```
